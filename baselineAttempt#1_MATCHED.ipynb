```python
# import stuff
import pandas as pd
import numpy as np
```

```python
# read the dataset file
circa_df = pd.read_excel("/content/drive/MyDrive/DL Project/Data/circa_test.xlsx")
```

```python
df = circa_df
```

```python
df['goldstandard1'] = df['goldstandard1'].astype(str)
df['goldstandard2'] = df['goldstandard2'].astype(str)
```

```python
df = df[(df['goldstandard1'] != 'nan')]
df = df[(df['goldstandard2'] != 'nan')]
```

```python
selected_columns = df[["id","goldstandard1","goldstandard2"]]
baseline_df = selected_columns.copy()
```

```python
baseline_df['majorityFn'] = 'Yes'
```

```python
from sklearn.metrics import classification_report
print(classification_report(baseline_df['goldstandard1'], baseline_df['majorityFn']))
```

```python
print(classification_report(baseline_df['goldstandard2'], baseline_df['majorityFn']))
```
